## Exploratory Data Analysis (EDA) Phase


```python
import matplotlib.pyplot as plt

plt.plot(df_filtered['date'], df_filtered['total_cases'])
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend('Total Cases over Time')
plt.show()
```

```python
plt.plot(df_filtered['date'], df_filtered['total_deaths'])
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.legend('Total Deaths over Time')
plt.show()
```

```python
plt.bar(df_filtered['date'], df_filtered['new_cases'], width = 10, linewidth=1)
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend('New Cases over Time')
plt.show()
```

```python
plt.bar(df_filtered['date'], (df_filtered['total_deaths']/df_filtered['total_cases']), width=12)
plt.xlabel('Date')
plt.ylabel('Total Deaths to Total Cases')
plt.legend('Deaths % over Time')
plt.show()
```

## Visualizing Vaccination Progress Phase

```python

plt.plot(df_filtered['date'], df_filtered['total_vaccinations'])
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.legend('Total Vaccinations over Time')
plt.show()
```

```python

plt.plot(df_filtered['date'], (df_filtered['people_vaccinated']/df_filtered['population']))
plt.xlabel('Date')
plt.ylabel('Percentage of Vaccinationed People')
plt.legend('Percentage of Vaccinationed People over Time')
plt.show()
```

```python
plt.plot(df_filtered['population'] , df_filtered['people_vaccinated'])
plt.xlabel('Population')
plt.ylabel('People Vaccinationed')
plt.legend('People Vaccinationed to the Overall Population')
plt.show()
```

```python
plt.hist((df_filtered['people_vaccinated']/df_filtered['population']))
plt.xlabel('People Vaccinationed to the Overall Population')
plt.show()
```

```python
df_filtered_USA = df_filtered[df_filtered['location'] == 'United States']
df_filtered_Pakistan = df_filtered[df_filtered['location'] == 'Pakistan']
df_filtered_India = df_filtered[df_filtered['location'] == 'India']

print('Total Vaccinations Over Time in USA')
plt.plot(df_filtered_USA['date'], df_filtered_USA['total_vaccinations'])
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.legend('Total Vaccinations Over Time in USA')
plt.show()

print('Total Vaccinations Over Time in Pakistan')
plt.plot(df_filtered_Pakistan['date'], df_filtered_Pakistan['total_vaccinations'])
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.legend('Total Vaccinations Over Time in Pakistan')
plt.show()

print('Total Vaccinations Over Time in India')
plt.plot(df_filtered_India['date'], df_filtered_India['total_vaccinations'])
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.legend('Total Vaccinations Over Time in India')
plt.show()
```

- United States was the first country to use vaccine and people start taking it with steady rate.
- Then India comes in the second place (from the selected countries) with somehow steady rate too.
- Pakistan only started taking the vaccine after 4 months from the other two countries, and even with unsteady rate!


```python
print(' Percentage of people_vaccinated to the Population in USA')
plt.hist((df_filtered_USA['people_vaccinated']/df_filtered_USA['population']))
plt.xlabel('% of people_vaccinated to the Population')
plt.show()

print('\n')
print(' Percentage of people_vaccinated to the Population in Pakistan')
plt.hist((df_filtered_Pakistan['people_vaccinated']/df_filtered_Pakistan['population']))
plt.xlabel('% of people_vaccinated to the Population')
plt.show()

print('\n')
print('Percentage of people_vaccinated to the Population in India')
plt.hist((df_filtered_India['people_vaccinated']/df_filtered_India['population']))
plt.xlabel('% of people_vaccinated to the Population')
plt.show()
```

- United States reached higher percentage in the vaccinated people over the total population in the country.
- People seems to rush at first in taking the vaccine, then slowed down, then back again on taking it on steady rate.
- There was a high peak in Pakistan, seems like there's a campain or something that encouraged people on taking the vaccine, then it slow down again.
- For India, seems like mostly they waited until they became sure of the safety of the vaccine after the rumors spread back at that time about various pharmutical compaines that provided the vaccines.


```python
print('Total Deaths To Total Cases Over Time in USA')
plt.bar(df_filtered_USA['date'], (df_filtered_USA['total_deaths']/df_filtered_USA['total_cases']), width=12)
plt.xlabel('Date')
plt.ylabel('Total Deaths To Total Cases')
plt.legend('Total Deaths To Total Cases Over Time in USA')
plt.show()


print('\n')
print('Total Deaths To Total Cases Over Time in Pakistan')
plt.bar(df_filtered_Pakistan['date'], (df_filtered_Pakistan['total_deaths']/df_filtered_Pakistan['total_cases']), width=12)
plt.xlabel('Date')
plt.ylabel('Total Deaths To Total Cases')
plt.legend('Total Deaths To Total Cases Over Time in Pakistan')
plt.show()


print('\n')
print('Total Deaths To Total Cases Over Time in India')
plt.bar(df_filtered_India['date'], (df_filtered_India['total_deaths']/df_filtered_India['total_cases']), width=12)
plt.xlabel('Date')
plt.ylabel('Total Deaths To Total Cases')
plt.legend('Total Deaths To Total Cases Over Time in India')
plt.show()
```

- The percentage of the total deaths to the total cases really got affected after taking the vaccines in both United States and India but for Pakistan; beacause of the very low rate and percentage that took the vaccines, the deaths rate hadn't decreased like the other two countries.
